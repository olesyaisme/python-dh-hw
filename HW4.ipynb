```python
f = open('Harry_Potter_and_the_Sorcerers_Stone.txt', 'r', encoding = 'utf-8')
text = f.read()
with open('hp.txt', 'w') as f:
    for item in text:
        f.write("%s\n" % item)
        
my_file = open('hp.txt', 'r', encoding = 'utf-8')
hp = my_file.read
```

```python
import nltk
from nltk.tokenize import word_tokenize
tokens = word_tokenize(text)
words = [word.lower() for word in tokens if word.isalpha()]
```

```python
from nltk import WordNetLemmatizer
wnl = WordNetLemmatizer()
lemmas = [wnl.lemmatize(w) for w in words]
```

```python
with open('lemmas.txt', 'w') as f:
    for item in lemmas:
        f.write("%s\n" % item)
```

```python
from collections import Counter 
counts = Counter(lemmas)
```

```python
sorted_counts = sorted(counts.items(), key=lambda kv: kv[1], reverse = True)
```

```python
b = {}
for lemma, count in counts.items():
    b[lemma] = (count/len(lemmas))
sorted_by_value = sorted(b.items(), key=lambda kv: kv[1], reverse = True)
```

```python
import matplotlib
import matplotlib.pyplot as plt

x = []
for k, v in enumerate(sorted_counts):
    x.append(k + 1)
y = []
for k, v in sorted_by_value:
    y.append(v)
```

```python
plt.plot(x, y)
plt.title("Закон Ципфа")
plt.xlabel("Ранг")
plt.ylabel("Относительная частота")
plt.show()
```

```python
from nltk.corpus import stopwords
sw = stopwords.words('english')
filtered = []
for word in lemmas:
    if word not in sw:
        filtered.append(word)
counts1 = Counter(filtered)
sorted_counts1 = sorted(counts1.items(), key=lambda x: x[1], reverse=True)
```

```python
x = []
y = list(range(30))

for k, v in (sorted_counts1[:30]):
    x.append(k)
plt.title("30 самых частотных слов")
plt.xlabel("Слова")
plt.ylabel("Ранг")
plt.bar(x, y)
plt.xticks(rotation=90)
plt.show()
```
